# Poisson statistics

```python
import numpy as np
import pylab as pl
```

## Check that the std of a Poisson distribution is sqrt(expectation)

```python
lam = 500
N = 30000
samples = np.random.poisson(lam=lam, size=N)
pl.hist(samples, bins=30)
print(np.std(samples))
print(np.sqrt(lam))
```

## comparison with the Gaussian approximation

```python
lams = [1, 10, 100, 1000]
N = 100000
for lam in lams:
    samples = np.random.poisson(lam=lam, size=N)
    pl.figure()
    bins_range = int(4*np.sqrt(lam)) + 1
    bins = np.arange(int(lam) - bins_range, int(lam) + bins_range + 1) + 0.5
    hist, _, _ = pl.hist(samples, bins=bins)
    x = np.linspace(min(bins), max(bins), 1000)
    gaussian_samples = np.random.standard_normal(N) * np.sqrt(lam) + lam
    hist, _, _ = pl.hist(gaussian_samples, bins=bins, alpha=0.5)
#print('check that the std of a Poisson distribution is sqrt(expectation)')
#print(np.std(samples))
#print(np.sqrt(lam))
```
