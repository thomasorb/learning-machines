## Ajustements (scipy.optimize)

```python
%matplotlib inline
import scipy
import numpy as np
import pylab as pl

def model(x, a, b):
    return a * np.exp(x/b)

x = np.arange(100, dtype=float)
A = 1
B = -15
y = model(x, A, B)


pl.plot(x,y)
```

```python
data = y + np.random.standard_normal(y.size) * 0.05
pl.plot(x, y)
pl.scatter(x, data, marker='+', color='orange')
```

https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.curve_fit.html
https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.leastsq.html

```python
import scipy.optimize
fit = scipy.optimize.curve_fit(model, x, data, 
                               method='lm',
                               p0=[0.,-1], # !! initial parameters must be given else they are all at zero
                               full_output=True) # levenberg-marquardt
print 'best fit', fit[0]
print 'covariance matrix'
print fit[1]
print 'other informations', fit[2].keys()

# how to compute an estimation of the uncertainty on the best fit 
# warning, this is true only if parameters are independant, i.e., off-diagonal covariance coefficient are near zero

print 'uncetainty on the parameters : ', np.sqrt(np.diag(fit[1])) 
pl.figure(figsize=(15,10))
pl.plot(x, y, label='real y')
pl.scatter(x, data, label='data', marker='+', color='orange')
pl.plot(x, model(x, *fit[0]), label='fitted model', color='red', ls=':')
pl.title('mieux')
pl.legend()
```
